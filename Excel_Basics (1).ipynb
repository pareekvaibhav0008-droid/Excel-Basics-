# Excel Basics


1.  Explain the difference between Relative , Absolute and Mixed Cell Referencing.

 ->  Relative - This is the Default type. When you copy the formula to another cell, the reference changes based on the relative position of rows and columns.
  
    Absolute - This uses dollar signs to lock both the row and column. The reference remains constant, no matter where the formula is copied

    Mixed - This Locks either only to row or only column, While the other remains relative.

7.Create a column chart showing total sales by product.

-> To create the column chart, follow these steps:
1. First, create a "Total Sales" column (Column G) using the formula =E2*F2.
2. Select the Product column (D1:D51) and the Total Sales column (G1:G51).
3. Click on Insert > Chart.
4. In the Chart Editor, select Column Chart as the chart type. This will show total sales for each product category.





8.Insert a line chart showing daily sales trend.

-> To create a Line Chart showing the daily sales trend, follow these steps:
1. Select Data: Highlight the OrderDate column (H1:H51) and the calculated Total Sales column (G1:G51).
2. Insert Chart: Go to the top menu and click on Insert > Chart.
3. Configure Setup: * In the Chart Editor, set the Chart type to Line chart.
* Set X-Axis to OrderDate.
* Set Series to Total Sales.
4. Result: This chart will visualize how sales fluctuate day-by-day across the entire dataset.
